```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Paths
DATA_PATH = Path("../../data/raw/BrentOilPrices.csv")
```

```python
# Load data
df = pd.read_csv(DATA_PATH)
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date').reset_index(drop=True)
```

```python
# Compute log returns
df['Log_Returns'] = np.log(df['Price']).diff()
```

```python

# Save for later use
LOG_PATH=Path("../../data/interim/brent_oil_log_returns.csv")
df.to_csv(LOG_PATH, index=False)
```

```python
# Plot raw price
plt.figure(figsize=(12,4))
plt.plot(df['Date'], df['Price'])
plt.title("Brent Oil Price Over Time")
plt.xlabel("Date"); plt.ylabel("Price ($)")
plt.grid(True)
plt.show()
```

```python
# Plot log returns
plt.figure(figsize=(12,4))
plt.plot(df['Date'], df['Log_Returns'])
plt.title("Log Returns of Brent Oil Price")
plt.xlabel("Date"); plt.ylabel("Log Return")
plt.grid(True)
plt.show()
```
